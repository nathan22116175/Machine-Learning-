```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# on filtre nos donnees
titanic = titanic[['survived', 'pclass', 'sex', 'age']]
titanic.dropna(axis = 0, inplace = True) #Elimine le donnee manquant dans le dataset
titanic['sex'].replace(['male', 'female'], [0, 1], inplace = True) #note : pour faire du machine learning on utilise des donner numerique pas des mots
titanic.head()
```

```python
y = titanic['survived']
X = titanic.drop('survived', axis = 1)
```

Séparation des données d'entraînement et de test :

```python
# Diviser les données en ensemble d'entraînement et ensemble de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

 Entraînement du modèle de régression logistique :

```python
# Initialiser et entraîner le modèle de régression logistique
model = LogisticRegression()
model.fit(X_train, y_train)
```

6. Évaluation du modèle :

```python
# Prédictions sur l'ensemble de test
y_pred = model.predict(X_test)

# Évaluer la précision du modèle
accuracy = accuracy_score(y_test, y_pred)
print(f"Précision du modèle : {accuracy * 100:.2f}%")
```

7. Prédiction pour un nouvel individu :

```python
# Prédiction pour un nouvel individu (par exemple, pclass=3, sex=0 pour un homme, age=26)
def survie(model, pclass=3, sex=0, age=26):
    n = np.array([pclass, sex, age]).reshape(1, -1)  # Redimensionner pour correspondre à l'entrée du modèle
    prediction = model.predict(n)  # Prédiction du modèle
    print("Survécu" if prediction[0] == 1 else "Non survécu")
```

```python
# Exemple de prédiction pour un homme de classe 3, âgé de 26 ans
survie(model)
```

- Régression logistique : Modèle basé sur une fonction logistique qui attribue une probabilité à chaque classe (0 ou 1) et fait une prédiction en fonction de cette probabilité.
- KNN : Modèle basé sur la distance entre les points pour prédire la classe d'un nouvel individu en se basant sur la majorité des voisins les plus proches.
