When you run this code, you'll see the model parameters (w and b) being updated at each step. It also prints the loss every 10 epochs to show how it's decreasing over time.

```python
import numpy as np
import matplotlib.pyplot as plt

# Dataset
X = np.array([1, 2, 3])  # Feature (x)
y = np.array([2, 3, 4])  # Target (y)

# Hyperparameters
eta = 0.1  # Learning rate
epochs = 100  # Number of iterations

# Initialize weights
w = 0
b = 0

# Function to compute the prediction
def predict(x, w, b):
    return w * x + b

# Function to compute the loss (Mean Squared Error)
def compute_loss(X, y, w, b):
    predictions = predict(X, w, b)
    return np.mean((y - predictions) ** 2)

# List to store loss values for plotting
losses = []

# Batch Gradient Descent (BGD)
for epoch in range(epochs):
    # Compute predictions for the entire dataset
    y_pred = predict(X, w, b)

    # Compute the gradients for w and b (using all data points)
    dw = -2 * np.dot(X, (y - y_pred)) / len(X)
    db = -2 * np.sum(y - y_pred) / len(X)

    # Update weights
    w = w - eta * dw
    b = b - eta * db

    # Optionally, print the loss at each epoch
    loss = compute_loss(X, y, w, b)
    losses.append(loss)

    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.4f}, w: {w:.4f}, b: {b:.4f}")

# Final parameters after BGD
print(f"Final parameters after {epochs} epochs: w = {w:.4f}, b = {b:.4f}")

# Plotting the loss curve
plt.plot(range(epochs), losses, label='Loss over epochs', color='blue')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss Curve for Batch Gradient Descent')
plt.legend()
plt.show()

# Plotting the data points and the line of best fit
plt.scatter(X, y, color='red', label='Data points')
plt.plot(X, predict(X, w, b), color='green', label='Line of best fit')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression using Batch Gradient Descent')
plt.legend()
plt.show()
```
